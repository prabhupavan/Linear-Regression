```python
x = data.drop('quality', axis=1)
y = data['quality']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.05)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(max_leaf_nodes = 7, min_samples_leaf = 5, max_depth= 4)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, classification_report

print(f'Accuracy: {accuracy_score(y_test.round(), y_pred.round())}')
```

```python
from sklearn import tree
fig = plt.figure(figsize=(20, 10))
tree.plot_tree(model,
               filled=True,
               fontsize=14)
```

```python
for i in range(len(y)):
    if(y[i]>= 7):
        y[i] = 'Good' 
    else:
        y[i] = 'Bad'
```

```python
y
```

```python
x_class_train, x_class_test, y_class_train, y_class_test = train_test_split(x, y, test_size=0.25)
```

```python
from sklearn.tree import DecisionTreeClassifier
class_model = DecisionTreeClassifier(max_leaf_nodes = 7, min_samples_leaf = 5, max_depth= 4)
class_model.fit(x_class_train, y_class_train)
```

```python
y_class_pred = class_model.predict(x_class_test)
y_class_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, classification_report

print(f'Accuracy: {accuracy_score(y_class_test, y_class_pred)}')
print(f'confusion_matrix:\n {confusion_matrix(y_class_test, y_class_pred)}')
print(f'precision_score: {precision_score(y_class_test, y_class_pred, average="micro")}')
print(f'recall_score: {recall_score(y_class_test, y_class_pred, average="micro")}')
print(f"classification_report:\n {classification_report(y_class_test, y_class_pred)}")
```

```python
from sklearn import tree
fig = plt.figure(figsize=(30, 10))
tree.plot_tree(class_model, class_names={0: 'Bad', 1:'Good'},
               filled=True,
               fontsize=14)
```
